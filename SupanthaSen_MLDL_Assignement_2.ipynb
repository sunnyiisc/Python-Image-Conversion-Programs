####  Q.1 write python code  from scratch to convert color image (Rose.jpg and Lenna.tif)  to grayscale image using  a def function and list of comprehension method and plot processing time comparison with  OpenCV and for loop method

```python
import numpy as npy
import cv2
from matplotlib import image, pyplot as plt
import time as ti
import math

def color2gray(img):
    #Image Dimensions
    img_row = img.shape[0]
    img_col = img.shape[1]

    #Image Band spliting (RGB Image)
    img_r = img[:,:,0]
    img_g = img[:,:,1]
    img_b = img[:,:,2]

    img_gray = npy.ones( shape=[img_row, img_col], dtype=img.dtype)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img_gray[i][j] = (0.2989*img_r[i][j]) + (0.5780*img_g[i][j]) + (0.1140*img_b[i][j])

    return(img_gray)


'''Lena Image Computation'''
img_lena = image.imread("lenna.tif")    #RGB reading of Image

plt.imshow(img_lena)
plt.title('Input Color Image')
plt.show()

#Method-1 using loop method
t_initial = ti.time()
img_lena_gray_loop = color2gray(img_lena)
t_final = ti.time()

time_lena_loop = t_final - t_initial

plt.imshow(img_lena_gray_loop, cmap='gray')
plt.title("Gray Lena Image from Loop Method")
plt.show()

#Method-2 using openCV method
t_initial = ti.time()
img_lena_gray_cv2 = cv2.cvtColor(img_lena, cv2.COLOR_BGR2GRAY)
t_final = ti.time()

time_lena_cv2 = t_final - t_initial

plt.imshow(img_lena_gray_cv2, cmap='gray')
plt.title("Gray Lena Image from cv2 Function")
plt.show()

print("Computation time for Loop Method :", time_lena_loop,'s')
print("Computation time for cv2 Method :", time_lena_cv2,'s')


'''Rose Image Computation'''
img_rose = image.imread("Rose.jpg")     #RGB reading of Image

plt.imshow(img_rose)
plt.title('Input Color Image')
plt.show()

#Method-1 using loop method
t_initial = ti.time()
img_rose_gray_loop = color2gray(img_rose)
t_final = ti.time()

time_rose_loop = t_final - t_initial

plt.imshow(img_rose_gray_loop, cmap='gray')
plt.title("Gray Rose Image from Loop Method")
plt.show()

#Method-2 using openCV method
t_initial = ti.time()
img_rose_gray_cv2 = cv2.cvtColor(img_rose, cv2.COLOR_BGR2GRAY)
t_final = ti.time()

time_rose_cv2 = t_final - t_initial

plt.imshow(img_rose_gray_cv2, cmap='gray')
plt.title("Gray Rose Image from cv2 Function")
plt.show()

print("Computation time for Loop Method :", time_rose_loop,'s')
print("Computation time for cv2 Method :", time_rose_cv2,'s')


'''Plotting the Time of Computation'''
barWidth = 0.25
lena = [time_lena_loop, time_lena_cv2]
rose = [time_rose_loop, time_rose_cv2]

br1 = npy.arange(len(lena))
br2 = [x + barWidth for x in br1]

plt.bar(br1, lena, color ='r', width = barWidth,
        edgecolor ='grey', label ='Lena Image')
plt.bar(br2, rose, color ='b', width = barWidth,
        edgecolor ='grey', label ='Rose Image')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (Seconds)')
plt.title('Comparision of Processing time (Absolute Value)')
plt.xticks([r + barWidth/2 for r in range(len(lena))],
        ['Loop Method', 'CV2 Method'])
plt.legend()
plt.show()

print('\nPlotting the log value of Processing time to make both bars visible in same plot:')
lena = npy.log10(lena)
rose = npy.log10(rose)

plt.bar(br1, lena, color ='r', width = barWidth,
        edgecolor ='grey', label ='Lena Image')
plt.bar(br2, rose, color ='b', width = barWidth,
        edgecolor ='grey', label ='Rose Image')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (log value of Seconds)')
plt.title('Comparision of Processing time (log Value)')
plt.xticks([r + barWidth/2 for r in range(len(lena))],
        ['Loop Method', 'CV2 Method'])
plt.legend()
plt.show()
```

####   Q.2. Write python code  from scratch to convert Grayscale  image to Binary images using a def function and list of comprehension method and  plot processing time comparison with OpenCV and for loop methods

```python
import numpy as npy
import cv2
from matplotlib import image, pyplot as plt
import time as ti

def gray2binary(img):
    img_row, img_col = img.shape

    img_bin = npy.ones( shape=[img_row, img_col], dtype=img.dtype )
    for i in range(img_row):
        for j in range(img_col):
            if img[i][j] > 127:
                img_bin[i][j] = 255
            else:
                img_bin[i][j] = 0
    return(img_bin)


img = cv2.imread('lenna.tif', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title('Input Grayscale Image')
plt.show()

#Method-1 using loop method
t_initial = ti.time()
bin_img_loop = gray2binary(img)
t_final = ti.time()

time_loop = t_final - t_initial

#Method-2 using cv2 method
t_initial = ti.time()
bin_img_cv2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
t_final = ti.time()

time_cv2 = t_final - t_initial

print("Computation time for Loop Method :", time_loop,'s')
print("Computation time for cv2 Method :", time_cv2,'s')

plt.subplot(1,2,1)
plt.imshow(bin_img_loop, cmap='gray')
plt.title('Binary Image using loop Method')
plt.subplot(1,2,2)
plt.imshow(bin_img_cv2, cmap='gray')
plt.title('Binary Image using cv2 Method')
plt.show()

'''Plotting the Time of Computation'''
barWidth = 0.5
comp_time = [time_loop, time_cv2]

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (Seconds)')
plt.title('Comparision of Processing time (Absolute Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()

print('\nPlotting the log value of Processing time to make both bars visible in same plot:')
comp_time = npy.log10(comp_time)

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (log value of Seconds)')
plt.title('Comparision of Processing time (log Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()
```

#### Q3.  Write python code from scratch to  add two images (image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and for loop methods.

```python
import numpy as npy
import cv2
from matplotlib import image, pyplot as plt
import time as ti


def image_add(img1, img2):
    
    actual_dtype = img1.dtype
    
    #Converting the Datatype to avoid Runtime Error
    img1 = img1.astype('int64')
    img2 = img2.astype('int64')
    
    img_row, img_col, band = img1.shape
    
    img_add = npy.ones( shape=[img_row, img_col, band], dtype=actual_dtype )
    for b in range(band):
        for i in range(img_row):
            for j in range(img_col):
                if (img1[i][j][b]+img2[i][j][b]) > 255:
                    img_add[i][j][b] = 255
                else:
                    img_add[i][j][b] = img1[i][j][b] + img2[i][j][b]
                    
    return(img_add)


img1 = image.imread("input1.jpg")    #RGB reading of Image
img2 = image.imread("input2.jpg")    #RGB reading of Image

plt.subplot(1,2,1)
plt.imshow(img1)
plt.title('Input Image 1')
plt.subplot(1,2,2)
plt.imshow(img2)
plt.title('Input Image 2')
plt.show()

#Method-1 using loop method
t_initial = ti.time()
img_add_loop = image_add(img1, img2)
t_final = ti.time()

time_loop = t_final - t_initial

#Method-2 using cv2 method
t_initial = ti.time()
img_add_cv2 = cv2.add(img1, img2)
t_final = ti.time()

time_cv2 = t_final - t_initial

print("Computation time for Loop Method :", time_loop,'s')
print("Computation time for cv2 Method :", time_cv2,'s')

plt.subplot(1,2,1)
plt.imshow(img_add_loop, cmap='gray')
plt.title('Added Image using loop Method')
plt.subplot(1,2,2)
plt.imshow(img_add_cv2, cmap='gray')
plt.title('Added Image using cv2 Method')
plt.show()

'''Plotting the Time of Computation'''
barWidth = 0.5
comp_time = [time_loop, time_cv2]

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (Seconds)')
plt.title('Comparision of Processing time (Absolute Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()

print('\nPlotting the log value of Processing time to make both bars visible in same plot:')
comp_time = npy.log10(comp_time)

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (log value of Seconds)')
plt.title('Comparision of Processing time (log Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()
```

#### Q.4 Write python code from scratch to subtract two images (image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and for loop methods.
 

```python
import numpy as npy
import cv2
from matplotlib import image, pyplot as plt
import time as ti


def image_sub(img1, img2):
    actual_dtype = img1.dtype
    
    #Converting the Datatype to avoid Runtime Error
    img1 = img1.astype('int64')
    img2 = img2.astype('int64')
    
    img_row, img_col, band = img1.shape
    
    img_sub = npy.ones( shape=[img_row, img_col, band], dtype=actual_dtype )
    for b in range(band):
        for i in range(img_row):
            for j in range(img_col):
                if (img1[i][j][b]-img2[i][j][b]) < 0:
                    img_sub[i][j][b] = 0
                else:
                    img_sub[i][j][b] = img1[i][j][b] - img2[i][j][b]
                    
    return(img_sub)


img1 = image.imread("input1.jpg")    #RGB reading of Image
img2 = image.imread("input2.jpg")    #RGB reading of Image

plt.subplot(1,2,1)
plt.imshow(img1)
plt.title('Input Image 1')
plt.subplot(1,2,2)
plt.imshow(img2)
plt.title('Input Image 2')
plt.show()

#Method-1 using loop method
t_initial = ti.time()
img_sub_loop = image_sub(img1, img2)
t_final = ti.time()

time_loop = t_final - t_initial

#Method-2 using cv2 method
t_initial = ti.time()
img_sub_cv2 = cv2.subtract(img1, img2)
t_final = ti.time()

time_cv2 = t_final - t_initial

print("Computation time for Loop Method :", time_loop,'s')
print("Computation time for cv2 Method :", time_cv2,'s')

plt.subplot(1,2,1)
plt.imshow(img_sub_loop, cmap='gray')
plt.title('Subtracted Image using loop Method')
plt.subplot(1,2,2)
plt.imshow(img_sub_cv2, cmap='gray')
plt.title('Subtracted Image using cv2 Method')
plt.show()

'''Plotting the Time of Computation'''
barWidth = 0.5
comp_time = [time_loop, time_cv2]

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (Seconds)')
plt.title('Comparision of Processing time (Absolute Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()

print('\nPlotting the log value of Processing time to make both bars visible in same plot:')
comp_time = npy.log10(comp_time)

br1 = npy.arange(len(comp_time))

plt.bar(br1, comp_time, color ='b', width = barWidth,
        edgecolor ='grey')

plt.xlabel('Method of Processing')
plt.ylabel('Processing Time (log value of Seconds)')
plt.title('Comparision of Processing time (log Value)')
plt.xticks([r for r in range(len(comp_time))],
        ['Loop Method', 'CV2 Method'])
plt.show()
```

```python
#   Q5.2
col_continent = cnty['continent']
continents = col_continent.unique()

print("Q5.2")
print('List of Continents (All):\n',col_continent.tolist())
print('List of Continents:\n',continents)
```
